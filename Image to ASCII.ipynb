# Image to ASCII 

```python
symbols = ['@', '#', '$', '%', '?', '*', '+', ';', ':', ',', '.']
```

#### Import the `PIL` library

```python
from PIL import Image
```

#### Load an image

```python
img = Image.open('laptop.jpg')
img
```

#### Resizing the image

```python
ratio = img.height / img.width / 2
new_width = 400
new_height = int(new_width * ratio)
res_img = img.resize((new_width, new_height))
res_img
```

#### Turn the image to grayscale (Black & White)

```python
res_img = res_img.convert('L')
res_img
```

#### Get the pixel data from the image using the `getdata` method

```python
all_pixels = res_img.getdata()
```

#### Now we can create a list of ASCII symbols and convert each pixel into a corrosponding symbol

```python
symbols_str = ''.join([symbols[pixel // 25] for pixel in all_pixels])
```

```python
ascii_img = '\n'.join([symbols_str[i:i+res_img.width] for i in range(0, len(all_pixels),res_img.width)])
```

```python
with open('ascii_img.txt', 'w') as text_img:
    text_img.write(ascii_img)
```
