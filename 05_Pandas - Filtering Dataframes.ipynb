Import libraries

```python
import pandas as pd
```

Import Data

```python
df = pd.read_csv('athlete_events.csv')
```

### Selecting Specific Rows

```python
df.index
```

### loc
By default, labels we specify refer to columns. If we want to refer to index labels we can use the ***loc*** command. Selecting a single row will return a Series with all columns as index labels

```python
df.loc[1]
```

```python
df.loc[2:4]
```

```python
df.loc[[1,3,4,9]]
```

### iloc
We can use "iloc" to specify the index position of the row

```python
df.iloc[0]
```

```python
df.iloc[-1]
```

```python
df.iloc[-7:-3]
```

## Dataframe Filtering
We can use conditional filtering to select rows that meet
certain criteria. To do that we need to generate a boolean Series using a conditional statement. In this example we are only interested in records where the athlete's age is greater than 85

```python
df['Age'].head() > 85
```

```python
age_filter = df['Age'] > 85
df[age_filter]
```

```python
df[df['NOC'] == 'TUV']
```
