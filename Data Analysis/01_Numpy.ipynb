# What is NumPy?
"__Numpy__ is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays. The ancestor of NumPy, Numeric, was originally created by Jim Hugunin with contributions from several other developers. In 2005, Travis Oliphant created NumPy by incorporating features of the competing Numarray into Numeric, with extensive modifications. NumPy is open-source software and has many contributors.


The Python programming language was not initially designed for numerical computing, but attracted the attention of the scientific and engineering community early on, so that a special interest group called matrix-sig was founded in 1995 with the aim of defining an array computing package. Among its members was Python designer and maintainer Guido van Rossum, who implemented extensions to Python's syntax (in particular the indexing syntax) to make array computing easier." 

_~ Wikipedia_




### *NumPy*, which stands for *Numerical Python*, is a library consisting of:
- Multidimensional array objects 
- A collection of routines for processing those arrays. 
- Mathematical and logical operations.




## Array Creation
The main object type in Numpy is called NDArry (Stands for N-Dimentional Array).<br>You can think about arrays as a much more sophisticated and efficient type of a list

```python
import numpy as np

my_list = [0,1,2,3,4]

arr = np.array(my_list)

print(arr)
```

numpy arrays allows us to perform operation on each member in the array without using loops

```python
print(my_list * 2)
print(arr * 2)
```

Unlike lists, each such array consists of a single data-type valeus (unless explicitly specified otherwise)

```python
arr.dtype
```

### arange
Creates a sequence of numbers and returns them as array. Takes in start,stop, and step size.
When specifying only a single parameter, numpy will treat it as the end point. Default range starts at 0

```python
numbers = np.arange(10)
print(numbers)
```

We can add a 2nd parameter to specify our starting point

```python
numbers = np.arange(5,16)
print(numbers)
```

3rd parameter will specift step size.<br> This example will start at 5 and print every other value (skiping every 2nd value)

```python
numbers = np.arange(5,16,2)
print(numbers)
```

## Numpy Arrays Indexing
### Slicing 1D Array

```python
num_vec = np.random.randint(10, 100, 10)
print(num_vec)
```

Selecting a specific value

```python
num_vec[3]
```

Selecting a range of values

```python
print(num_vec[3:8])

print(num_vec[3:])

print(num_vec[:8])
```

Selecting multiple, non consecutive values (note the double brackets)

```python
print(num_vec)
print(num_vec[[2,9,4,4]])
```

We can change one value

```python
num_vec[3] = 999
print(num_vec)
```

But we can also change all values in a slice at once

```python
num_vec[3:] = 999
print(num_vec)
```

We can put a slice inside another variable

```python
vec_slice = num_vec[3:]
```

This slice will serve as a view of the original array.
Any change to it will reflect on the original array as well

```python
vec_slice[:] = 0
print(vec_slice)
print(num_vec)
```

This was made to save memory when dealing with large arrays
but we can overide this behaviour by using the copy method

```python
vec_slice = num_vec.copy()[3:]
print(vec_slice)
vec_slice[:] = 333
print(vec_slice)
print(num_vec)
```

## Filtering Arrays

The masking technique allows to use conditional filters to get only certain elements from an array

```python
payments = np.random.uniform(100, 1000, (20, 4)).round(2)
print(payments)
```

```python
mask = payments > 450.50
print(mask)
```

Apply the mask to the array

```python
print(payments[mask])
```

We can skip the use of that additional variable

```python
print(payments[payments > 450.50])
```

### multiple filters 
(& = and, | = or)

```python
mask1 = payments > 100
mask2 = payments < 300
print(payments[mask1 & mask2])
```

We can also write this in a single statement

```python
print(payments[(payments > 100) & (payments < 300)])
```

We can reverse the meaning of a filter using the `~` sign

```python
values = np.random.randint(0,9,(5,5))
values
```

```python
above_4 = values >= 4
above_4
```

Get all values below 4

```python
values[~above_4]
```

### where()
The ***where*** method functions like the `if` function in Excel - Define a condition and then the value you want to get when condition is met, and a third value to get otherwise

```python
np.where(above_4, '$', '###')
```

## numpy datatypes in depth


**String**

`S` - Followed by a number indicating it's length. (e.g. S8).

**Boolean:**

`bool`  - True or False. stored as a byte

**Integers:**

- `int` - Default integer type  (32 or 64)
- `int8` - Byte (-128 to 127)
- `int16` - Integer (-32768 to 32767)
- `int32` - Integer (-2147483648 to 2147483647)
- `int64` - Integer (-9223372036854775808 to 9223372036854775807)

**Unsigned Integers**

- `uint8` (0 to 255)
- `uint16` (0 to 65535) 
- `uint32` (0 to 4294967295)
- `uint64` (0 to 18446744073709551615)

**Float**

`float16` - Half precision float: sign bit, 5 bits exponent, 10 bits mantissa
`float32` - Single precision float: sign bit, 8 bits exponent, 23 bits mantissa
`float64` - Double precision float: sign bit, 11 bits exponent, 52 bits mantissa

**Complex number:**

`complex64` - Complex number, represented by two 32-bit floats (real and imaginary components)
`complex128` - Complex number, represented by two 64-bit floats (real and imaginary components)



### Using Data Type Objects (dtype)
A `data type object` describes the data types of the objects in the **ndarray**. 

Using data types objects helps us manage and access our objects more 'humanly' manner.

We can use ***np.dtype*** to define the data type of the object

```python
my_type = np.dtype(np.int16)
my_type
```

```python
my_type = np.dtype(np.int16)
np.array([1,2,3,4,100,200,300,400], dtype=my_type)
```

When working with two-dimentioanl datasets, we can form header-like titles for our "columns" and access these columns directly by specifying their name.

```python
sales_dt = np.dtype([('sales', np.float32)])

sales_data = np.array([(100),(200),(300),(400)], dtype=sales_dt)
sales_data
```

```python
sales_data['sales']
```

This technique makes more sense when dealing with multiple columns (notice the string type 

```python
order_dt = np.dtype([('customer','S20'), ('units', 'int8'), ('price', 'float16')]) # note the string data type
orders = np.array([('Elad', 21, 50.3),
                   ('Ram', 28, 75.0),
                   ('Dani', 22, 95.1),
                   ('Tal', 23, 79.0),
                   ('Yossi', 19, 82.9),
                  ], dtype = order_dt) 
print(orders)
```

```python
orders['customer']
```

```python
profit = orders['units'] * orders['price']
profit
```

### More Data Manipulation With NDArrays

```python
numbers = np.array([[1,2],[3,4]])
letters = np.array([['a','b'],['c','d']])
print(numbers)
print(letters)
```

### concatenate()
We can merge arrays together with the ***concatenate*** method

```python
characters = np.concatenate((numbers, letters))
characters
```

### axis
We can change the axis to combine them horizontaly instead of vertically

```python
characters = np.concatenate((numbers, letters), axis = 1)
characters
```

### append()
Appends the values to the end of an array. This will generate a 1d array

```python
print(np.append(characters,123))
print(np.append(characters,123).ndim)
```

If we want to keep the multi-dimentional structure of the array we can set the axis to 0 and add a "row" (we'll have to supply the right amount of values now)

```python
np.append(characters,[[11,22,33,44]], axis=0)
```
