## Dataframe Filtering
We can use conditional filtering to select rows that meet
certain criteria. To do that we need to generate a boolean Series using a conditional statement. In this example we are only interested in records where the athlete's age is greater than 85

```python
df = pd.read_csv('athlete_events.csv', index_col='ID')
df['Age'].head() > 85
```

```python
age_filter = df['Age'] > 85
df[age_filter]
```

We can also use the condition directly in the slicing brackets

```python
df[df['NOC'] == 'TUV']
```

### Applying Multiple Filters
We use the '&' to apply the **and** logic (all conditions must evaluate to true)

```python
filter_1 = df['NOC'] == 'STP'
filter_2 = df['City'] == 'Beijing'
filter_3 = df['Sex'] == 'M'

df[filter_1 & filter_2 & filter_3]
```

Again, we can skip the variables and write the conditions directly within the brackets. In cases of multiple conditions we'll have to use parentheses

```python
df[(df['NOC'] == 'STP') & (df['City'] == 'Beijing') & (df['Sex'] == 'M')]
```

We use the '|' ("pipe") to apply the **or** logic (at least one of the conditions must evaluate to true)

```python
filter1 = df['NOC'] == 'NBO'
filter2 = df['Event'] == 'Sailing Mixed 18 foot'

df[filter1 | filter2]
```

And in a single line:

```python
df[(df['NOC'] == 'NBO') | (df['Event'] == 'Sailing Mixed 18 foot')]
```

We can mix the two options

```python
filter1 = df['City'] == 'London'
filter2 = df['Age'] > 75
filter3 = df['Height'] < 128

df[(filter1 & filter2) | filter3]
```

We can use the '~' sign ("Tilda") to apply a **not** logic thus reversing the meaning of an expression.
In this example the **not** operation would apply only for filter1 and filter2 because of the parentheses

```python
filter1 = df['City'] == 'London'
filter2 = df['Age'] > 75
filter3 = df['Height'] < 128

df[~(filter1 & filter2) & filter3]
```

### isin()
We can use the **isin** method if we want to filter our Dataframe by several values as once

```python
teams = ['China', 'Denmark', 'Netherlands', 'Japan']
df['Team'].head(10).isin(teams)
```

```python
df[df['Team'].isin(teams)].head(7)
```

We can use the '~' sign here as well, to search for all the other teams

```python
df[~df['Team'].isin(teams)].head(7)
```

### between()
We can use the **between** method to look for a range of values

```python
df[df['Height'].between(160, 180)].head(5)
```

```python
df[~df['Height'].between(160, 180)].head(5)
```

### Filtering empty values

```python
df[['Name', 'Team', 'City', 'Medal']].head(7)
```

### isnull()
The **isnull** method returns True for every missing (NaN) value

```python
df[['Name', 'Team', 'City', 'Medal']].head(7).isnull()
```

Use it to filter our Dataframe - display only athletes without medals

```python
df[df['Medal'].isnull()][['Name', 'Team', 'City', 'Medal']].head(7)
```

### notnull()
The **notnull** does the exact opposite and returns True for actual values (not missing)

```python
df[df['Medal'].notnull()][['Name', 'Team', 'City', 'Medal']].head(7)
```

## Changing values based on a filter
Trying to sub-set a dataframe and change the result won't work

```python
football_players = df['Sport'] == 'Football'
df[football_players].head()
```

```python
df[football_players]['Sport'] = 'American Football'
```

The data didn't change, of course

```python
df[football_players].head(3)
```

The reason for this is because filtering creates a copy of the data, without saving the copy into a new variable, the change is meaningless, and even if we assign it to a new variable, the original dataframe won't be affected

```python
football_df = df[football_players]
```

```python
football_df['Sport'] = 'American Football'
```

```python
football_df.head()
```

```python
df[football_players].head()
```

The correct way to perform this operation will be to use the ***loc*** method. ***loc*** does not create a copy from the data and therefore any change we apply to this subset will actually affect the original dataframe

```python
df.loc[football_players, 'Sport'] = 'American Football'
df.head()
```

## where()
We can use the ***where*** method to keep all rows of the dataframe but show values only for selected rows

```python
df.where(df['NOC'] == 'DEN').head()
```

```python
df.where((df['NOC'] == 'DEN') & (df['Year'] == 1920)).head()
```

## query()
The ***query*** method is yet another way to filter the results we get from a dataframe in an arguably more convient syntax. Because we are specifying our column names in plain text inside the ***query***'s string, only columns without spaces in their names are viable here. Make sure to eliminate any spaces in advance.

```python
df.query('NOC == "DEN"').head()
```

```python
df.query('NOC == "DEN" & Year == 1900').head()
```

To make the writing even more friendly, we can use Python's traditional logic operators (and / or / not / in)

```python
df.query('NOC == "DEN" and Year == 1900').head()
```

```python
df.query('NOC not in ("DEN","CHN","NED") or Year == 1900').head()
```
