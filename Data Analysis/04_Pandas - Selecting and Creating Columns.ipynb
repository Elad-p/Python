Import Data

## Selecting Columns

### Order of Operations
*dataframe_name*[*column_name*] <br>
**or**<br>
*dataframe_name*[[*columns_names*]]
<br><br><br>
Selecting a single column will return a Series

```python
df['Name'].head()
```

To select multiple columns, we need to put them in a single list

```python
df[['Name', 'Sex', 'Age', 'Team']].head()
```

Individual columns can be selected with a dot after them. <br>Easier to write (we don't even need to include quotes) but more confusing (looks like a method) hence not recommended

```python
df.Name.head()
```

## sample()
The ***sample*** method will return a random record from the dataframe

```python
df.sample()
```

We can increase the number of random records being pulled by using the ***n*** parameter

```python
df.sample(3) # This is the equivalent of writing df.sample(n = 3)
```

We can even use ***sample*** to pull random columns by setting the ***axis*** parameter to 1

```python
df.sample(n = 3, axis = 1).head()
```

## Rename columns
The **rename** method accepts a dictionary with current column names as keys, and the names we want to change them into - as values

```python
df = pd.read_csv('athlete_events.csv', index_col = 'Name')
df.rename(columns={'Sex': 'Gender', 'Age': 'Age of Athlete', 
                   'Height': 'Height in cm', 'Weight': 'Weight in cm'}).head()
```

Use the **inplace** to save your changes

```python
df.rename(columns={'Sex': 'Gender', 'Age': 'Age of Athlete','NOC':'Country Code'}, inplace=True)
df.head()
```

We can also use ***rename*** to change row labels (index)

```python
df.rename(index = {'A Dijiang': 'Dijiang','A Lamusi': 'Lamusi', 'Gunnar Nielsen Aaby': 'Nielsen'},
         columns = {'Sex': 'Gender', 'Age': 'Age of Athlete','NOC':'Country Code'}, inplace = True)
df.head()
```

If we want to change all columns at once we can just change the ***columns*** attribute

```python
df.columns = ['ID', 'Gender', 'Age', 'Height', 'Weight', 'Team',
       'NOC', 'Games', 'Year', 'Season', 'City', 'Sport', 'Event',
       'Medal']
df.head()
```

## Adding & Removing Columns
Selecting a nonexistent column name will return a key error

```python
df = pd.read_csv('athlete_events.csv')
```

```python
df['BMI'] # Error!
```

Assigning a value to a nonexistent column, on the other hand, will generate it and add it to the dataframe permanantly. Columns generated in this manner would always be added on the far right

```python
df['BMI'] = df['Weight'] / (df['Height'] / 100) ** 2
df.head()
```

### insert
We can use the ***insert*** method to specify the position of our new column

```python
df.insert(6, 'Weight in Pounds', df['Weight'] * 2.2)
df.head()
```

### drop
If we want to delete new column we can specift that using drop. We also need to specify axis=1 because we can use drop to remove rows as well and that's the default

```python
df.drop('Weight in Pounds', axis=1).head()
```

We can drop an entire list of columns at once

```python
df.drop(['Games', 'Year', 'Season','Medal', 'City','Sport'], axis=1).head()
```

### inplace
The change of the dataframe won't stick unless we use the ***inplace*** parameter

```python
df.head(3)
```

```python
df.drop('Weight in Pounds', axis=1, inplace=True)
df.head(3)
```

We can delete rows in the same manner

```python
df.drop(1).head(3)
```

### pop()
Removing columns with the ***pop*** method (works only for columns). 

```python
df.pop('Weight').head()
```

The pop method returns back a column but also removes it permanantly from the data frame. This method can be useful for transfering columns between different dataframe objects

```python
df.head()
```

We can also use the general command ***del*** to discard a column

```python
del df['Height']
df.head()
```
