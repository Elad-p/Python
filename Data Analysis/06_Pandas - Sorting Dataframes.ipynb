# Sorting Dataframes
Import libraries

```python
import pandas as pd
```

Import data

```python
df = pd.read_csv('athlete_events.csv', index_col='ID')
df.head()
```

### sort_values()
The **sort_values** method allows us to sort the entire Dataframe by a specific column

```python
df[['Name', 'Sex', 'Age', 'Height', 'Sport', 'Team']].sort_values('Age').head(7)
```

We can use the **ascending** argument to reverse the sorting

```python
df[['Name', 'Sex', 'Age', 'Height', 'Sport', 'Team']].sort_values(by='Age', ascending=False).head(7)
```

### inplace
By default changes we make to our Dataframe object (like sorting the values) will only be calculated without actually affecting the object. We can change that using the **inplace** parameter

```python
df.sort_values(by='Age', inplace=True)
df[['Name', 'Sex', 'Age', 'Height', 'Sport', 'Team']].head(7)
```

The previous cell is the more "pandas way" equivelent of writing the following:

```python
df = df.sort_values(by='Age')
df[['Name', 'Sex', 'Age', 'Height', 'Sport', 'Team']].head(7)
```

### Secondary Sorting

```python
df.sort_values(by='Team', inplace=True)
df[['Name', 'Sex', 'Age', 'Height', 'Sport', 'Team', 'Medal']].head(9)
```

The **by** argument can accept a list of columns

```python
df.sort_values(by=['Team', 'Age'], inplace=True)
df[['Name', 'Sex', 'Age', 'Height', 'Sport', 'Team', 'Medal']].head(9)
```

For each of those columns we can decide if it should be ascending or descending by passing a corrosponding list to the **ascending** argument

```python
df.sort_values(by=['Team', 'Age'], inplace=True, ascending=[True, False])
df[['Name', 'Sex', 'Age', 'Height', 'Sport', 'Team', 'Medal']].head(9)
```

By default missing values would be listed at the end. We can change that with the **na_position** argumetn (notice row #3)

```python
df.sort_values(by=['Team', 'Age'], inplace=True, ascending=[True, False], na_position='first')
df[['Name', 'Sex', 'Age', 'Height', 'Sport', 'Team', 'Medal']].head(9)
```

### sort_index()
We can use the **sort_index** method to sort the Dataframe by it's index

```python
df.head()
```

```python
df.sort_index().head()
```

We can reverse the order when sorting by index as well

```python
df.sort_index(ascending=False).head()
```

Keep the change with the **inplace** argument

```python
df.sort_index(ascending=False, inplace=True)
df.head()
```

## Duplicate rows
The method **duplicated** will return False only for the 1st occurrence of each value

```python
df = pd.read_csv('athlete_events.csv', index_col='ID')
df.sort_values(by='Team', inplace=True)
```

```python
df['Team'].head(9)
```

```python
df['Team'].duplicated().head(9)
```

We can use it to filter out the first row of each team

```python
df[df['Team'].duplicated()].head(9)
```

### keep
We can mark the last of each group as unique, instead of the first (which is the default)

```python
df['Team'].duplicated(keep='last').head(9)
```

We can also mark as duplicate every value that repeats more then once overall

```python
df['Team'].duplicated(keep=False).head(9)
```

Select only names that appears once throughout the Dataframe

```python
df.sort_values(by='Name', inplace=True)
df['Name'].head(10)
```

```python
df['Name'].duplicated(keep=False).head(10)
```

```python
df[~df['Name'].duplicated(keep=False)].head(10)
```

### drop_duplicates()

```python
df[['NOC', 'Team']].sort_values('NOC').head(10)
```

```python
df[['NOC', 'Team']].sort_values('NOC').drop_duplicates().head(10)
```
